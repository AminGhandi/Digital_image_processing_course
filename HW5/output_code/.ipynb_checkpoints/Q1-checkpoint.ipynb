```python
def dilation(image_path, kernel_size):
    # Read the image for dilation
    img2 = image_path
    
    # Acquire size of the image
    p, q = img2.shape
    

    # Define new image to store the pixels of dilated image
    imgDilate = np.zeros((p, q), dtype=np.uint8)
    
    # Define the structuring element
    SED = np.ones((kernel_size, kernel_size), dtype=np.uint8)
    constant1 = (kernel_size - 1) // 2
    
    # Dilation operation without using inbuilt CV2 function
    for i in range(constant1, p - constant1):
        for j in range(constant1, q - constant1):
            temp = img2[i - constant1 : i + constant1 + 1, j - constant1 : j + constant1 + 1]
            product = temp * SED
            imgDilate[i, j] = np.max(product)
            
    # Return the dilated image without padding
    return imgDilate



def erosion(image_path, kernel_size):
    # Read the image for erosion
    img1 = image_path
    
    # Acquire size of the image
    m, n = img1.shape 
    

    
    # Define the structuring element
    SE = np.ones((kernel_size, kernel_size), dtype=np.uint8)
    constant = (kernel_size - 1) // 2
    
    # Define new image
    imgErode = np.zeros((m, n), dtype=np.uint8)
    
    # Erosion without using inbuilt cv2 function for morphology
    for i in range(constant, m - constant):
        for j in range(constant, n - constant):
            temp = img1[i - constant : i + constant + 1, j - constant : j + constant + 1]
            product = temp * SE
            imgErode[i, j] = np.min(product)
    
    # Return the eroded image without padding
    return imgErode

def opening(image, kernel_size):
    eroded = erosion(image, kernel_size)
    return dilation(eroded, kernel_size)

def closing(image, kernel_size):
    dilated = dilation(image, kernel_size)
    return erosion(dilated, kernel_size)
```

```python



# Load the images
sky1 = cv2.imread('q1/sky1.jpeg', 0)
sky2 = cv2.imread('q1/sky2.jpg', 0)
sky3 = cv2.imread('q1/sky3.jpg', 0)
sky4 = cv2.imread('q1/sky4.jpg', 0)

# Threshold value to select bright stars
threshold = 100

# Function to remove small stars using opening morphological operation
def remove_small_stars(image):
    opened=opening(image, 3)
    return opened.astype(np.uint8)

# Function to make stars convex using convex hull operation
def make_stars_convex(image):
    contours, _ = cv2.findContours(image, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    convex_image = np.zeros_like(image)
    for contour in contours:
        hull = cv2.convexHull(contour)
        cv2.drawContours(convex_image, [hull], 0, 255, -1)
    return convex_image

# Function to find the top n brightest stars
def find_brightest_stars(image, n):
    # Threshold the image to extract bright areas
    _, thresh = cv2.threshold(image, threshold, 255, cv2.THRESH_BINARY)

    # Find contours of the bright areas
    contours, _ = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    # Sort contours based on contour area (brightest stars have larger areas)
    sorted_contours = sorted(contours, key=lambda x: cv2.contourArea(x), reverse=True)

    # Extract the top n contours (brightest stars)
    top_contours = sorted_contours[:n]

    # Create a black image to draw the top contours
    star_image = np.zeros_like(image)

    # Draw the top contours on the black image
    cv2.drawContours(star_image, top_contours, -1, 255, -1)

    return star_image

# Remove small stars from each image
processed_sky1 = remove_small_stars(sky1)
processed_sky2 = remove_small_stars(sky2)
processed_sky3 = remove_small_stars(sky3)
processed_sky4 = remove_small_stars(sky4)


# Find and make the top 10 brightest stars convex in each image
convex_star_image1 = make_stars_convex(find_brightest_stars(processed_sky1, 10))
convex_star_image2 = make_stars_convex(find_brightest_stars(processed_sky2, 10))
convex_star_image3 = make_stars_convex(find_brightest_stars(processed_sky3, 10))
convex_star_image4 = make_stars_convex(find_brightest_stars(processed_sky4, 10))

# Display the results
cv2.imwrite('q1/Convex_Bright_Stars_Sky_1.jpg', convex_star_image1)
cv2.imwrite('q1/Convex_Bright_Stars_Sky_2.jpg', convex_star_image2)
cv2.imwrite('q1/Convex_Bright_Stars_Sky_3.jpg', convex_star_image3)
cv2.imwrite('q1/Convex_Bright_Stars_Sky_4.jpg', convex_star_image4)
```
